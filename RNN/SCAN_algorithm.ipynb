## 1. Introduction

** Network clustering ** 은 유전자, 컴퓨터 네트워크, graph로 표현되는 전반적인 데이터에서 사용되는 기술이다. <br>
본 논문은 undirected and un-weighted graph를 기반하여 설명한다. <br>

![network](https://i.imgur.com/8TaxwQz.png) 

목적은 다음과 같다 : <br>
* *hub(cluster 간 연결점)* : 위 그림의 (6) node
* *clustering* : 위 그림의 (0,1,2,3,4,5), (7,8,9,10,11,12)
* *outlier 제거* : 위 그림의 13

**network clustering**에 대한 유명한 다른 접근방식으로는 다음과 같은 것들이 있다 :
* Modularity-based algorithms
* Normalized cut

사실 *Modularity-based algorithm*은 network clustering에 있어 매우 성공적인 방식이지만, <br>
다음과 같은 한계가 있다 : 
* time complexity : O($md log n$), m = |E|, n = |V|, d = hyperparameter
* does not find hub & outlier

**SCAN algorithm** 은 *Modularity-based algorithm* 에 비하여 시간적, 기능적 장점이 있다.
* time complexity : O(m), m = |E|
* finds hub & outlier

## 2. Structure-connected clusters 란

위에서 언급한 network clustering은 
1. number of edges between vertices in a cluster
2. number of edges between clusters.

이 두 가지를 고려한다. <br>
하지만 본 방식은 *'direct connection'* 만을 고려한다는 한계가 있다.<br>

따라서 SCAN은 **'common neighbors'**를 고려한다. <br>
즉, vertices간의 neighbor를 비교하는 것이다. <br>
예를들어, 학교에서 mutual friend가 많은 친구들은 같은 집단에 속한다는 생각을 해보면 자명하다. <br>

따라서 '구조가 비슷한 것들끼리 모인 집합'(structure connected clusters) 을 정의한다.

## 3. Definitions.

프로세스는 다음과 같다. <br>
1. Core, 즉 cluster의 중심이 되는 vertex를 뽑는다.
2. core와 비슷한 structure로 연결된 cluster를 구한다.
3. cluster에 포함되지 않은 node를 hub와 outlier로 나눈다.

### 3.1 VERTEX STRUCTURE
#### 가장 먼저, neiborhood 집합을 정의한다.

$\Gamma(v) = \{w \in V | (v,w) \in E\} \cup \{v\}$ <br>

node $v$와 연결된 모든 vertices와 해당 node 자체를 먼저 neighborhood로 정의한다.

### 3.2 STRUCTURAL SIMILARITY

그 다음, 위에 설명한 대로, 친구가 비슷한 vertices들을 유사하다고 계산할 수 있는 <br>
**Structural similarity**를 정의한다.<br>

$\sigma (v,w) = \dfrac{|\Gamma(v) \cap \Gamma(w)|}{\sqrt{|\Gamma(v)| |\Gamma(w)|}}$

친구가 많을수록 같은 cluster라고 할수 없음으로, neiborhood의 수로 normalize를 해줘야 함은 자명하다.



### 3.3 $\epsilon$-neiborhood

붙어있다고 다 친구가 아니다. 따라서 일정한 기준을 통하여 cluster membership을 만드는 방식을 계산한다. <br>

$N_{\epsilon}(v) = \{ w \in \Gamma(v) | \sigma (v,w) \geq \epsilon \} $ <br>

$\epsilon$ 이라는 hyperparameter를 통하여, 일정 값이 넘는 pair를 $\epsilon$-neiborhood 라 칭한다. <br>

### 3.4 CORE

이제 **structural similarity**를 통하여 cluster의 중심이 되는 core vertices를 뽑는 법을 정의한다. <br>

여기서 핵심은, 하나의 cluster에는 하나의 core가 있는 것이 아니라는 점이다. <br>
